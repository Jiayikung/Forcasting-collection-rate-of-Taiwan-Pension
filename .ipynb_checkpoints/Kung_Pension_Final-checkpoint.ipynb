### Abstract

Taiwan's National Pension Insurance (NPI) is a compulsory defined benefit pension insurance scheme; however, the collection rate of NPI has fallen below 50% since January 2019.. This study aims to analyze the impact of four key macroeconomic factors on the collection rate of NPI and accordingly predict future trends based on one of these factors. Using time-series data from 2008 to 2021, the study employs a multiple regression model to determine the most influential macroeconomic variable on collection rates. Next, we utilize Bayesian estimation model to forecast future collection rates, providing valuable insights into future trends and potential outcome. The results indicate that the monthly economic growth rate is the most influential factor, with a 65% probability that higher economic growth leads to lower collection rates. These findings shed light on a negative relationship between economic growth and NPI collection rate, highlighting an inherent issue concerning the lack of engagement and motivation within the policy.

The study is as follows. Section 1 mentions background information and Literature Reviews. Section 2 describes the data and sample. Section 3 outlines the multiple regression model and the Bayesian estimation model. Section 4 concludes.

### 1.1 Background Information

Taiwan's population is aging, and as a result, society is paying more and more attention to the life insurance provided to retirees. Other insurances in Taiwan, however, have higher insured requirements, making it impossible for residents with incomes below the average to get insured. As a result, on October 1, 2008, the government of Taiwan passed National Pension Insurance. **National Pension Insurance (NPI)**, according to the Bureau of Labor Insurance, Ministry of Labor (2007), **aims to ensure the basic economic security for citizens who are unable to be adequately covered by related social insurance when they become old, pregnant, or suffer from a mental or physical disability, as well as the stability of their survivors' lives.**

As for some key points about Taiwan's National Pension Insurance (NPI), individuals aged 25-65 not covered by other social insurance must join NPI and pay the premium bi-monthly. Additionally, at 65, the insurance holders can recieve following benefits: old-age pension payments, maternity payments, pension payments for mental and physical disabilities, death payments, and pension payments for surviving family members. Last but not least, since the majority of insured people are unemployed or economically disadvantaged, the insured won't be penalized for late premiums; however, failure to pay premiums within ten years will result in loss of future benefits.

According to monthly data collected from the National Statistics Bureau, the insurance collection rate noticeably dropped from 77.1% in 2008 to 41.4% in 2021, as shown in Figure 1. Most importantly, Figure 2. reveals that 38% of those who did not consistently pay cited the inability to afford the monthly premium as the reason. Apparently, examining whether spurring economic growth could enhance NPI collection rate is an issue worth exploring.

### 1.2 Literature Review

Existing Research on Taiwan's National Pension Insurance (NPI) has widely mentioned two aspects -- studies on the comparison between different countries in terms of the pension insurance scheme, and studies on its impact on income redistribution and its mutual influences with several macroeconomic variables. 

For studies demonstrating the comparison, Hsu (2009) conducted a comparative analysis of the National Pension System in Japan, Germany, and Taiwan to understand profoundly the role and potential challenges of the system and propose suggestions to the government for future implementation. Using several theories to discuss current disputes about the legitimacy of the premium, Szu (2009) analyzed the content and problems of the five benefit payments.

On top of the issue mentioned above, most studies consider NPI cannot bring income redistribution and has a negative impact on the macroeconomic environment. Shie (2012) examined its influence on income redistribution, finding it doesn't achieve this goal due to its primary focus on providing a minimum standard of living for the non-income population in old age. Lu (1999) used a social accounting matrix (SAM) to assess the macroeconomic effects, showing results aligned with academic theories but offering unique insights into the social cost of NPI. Similarly, to support this perspective, Yu (2018) investigated the impact of five macroeconomic variables on the collection rate of National Pension Insurance (NPI). Using time-series data from 2008 to 2017, various analyses were performed, revealing a positive correlation between the collection rate and the unemployment rate, and negative correlations with average GDP per capita, savings rate, and the consumer price index (CPI). **Based on the above research, our research hypothesis was developed: The NPI's collection rate is negatively related to the macroeconomic variables.**

Considering the limited discussion on NPI's overall economic implications. Our study offers a novel contribution by exploring the impact of four macroeconomic variables on the collection rate, incorporating lagged effects and time trends. Additionally, we propose a forecast model for the NPI collection rate based on influential economic factors. This analysis seeks to provide policymakers with valuable insights into the future trend and the potential reasons behind it.

```python
# Use Tukey’s Fences Method to identify outliers 

# Create a new DataFrame to store the cleaned data
select_col_cleaned = pd.DataFrame()

# Identify rows with outliers
outliers_mask = np.zeros(len(select_col), dtype=bool)

for col in select_col.columns:
    # Calculate 25th and 75th percentiles and store in q1 and q3
    q1 = np.percentile(select_col[col], 25)
    q3 = np.percentile(select_col[col], 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    # Identify outliers for each column
    # True indicates an outlier, and False indicates a non-outlier.
    outliers_mask |= (select_col[col] < lower_fence) | (select_col[col] > upper_fence)

# Append the 'month' variable to `select_col_cleaned` DataFrame
# The 'month' variable is included in the cleaned DataFrame only for the rows that are not outliers
select_col_cleaned = df.loc[~outliers_mask].copy()
select_col_cleaned.reset_index(drop=True, inplace=True)
```

### 3.1 Multiple Regression model

I used the following multiple regression model to examine to what extent the decline of the collection rate of National Pension Insurance can be accounted for by macroeconomic-related changes after controlling the lagged effect, the time trend, and the year correlation.

$\hat{Pension}_{t} = \hat{\beta}_{0}+\hat{\beta}_{1}\times\text{CPI}_{t}+\hat{\beta}_{2}\times\text{Economic growth rate}_{t}+\hat{\beta}_{3}\times\text{GDP per capita}_{t}+\hat{\beta}_{4}\times\text{Unemployment rate}_{t}+\hat{\beta}_{5}\times\text{Year}+\hat{\beta}_{6}\times\text{Year}^{2}+\hat{\beta}_{7}\times\text{lag}+\text{ε}$

where the subscript $\text{t}$ represents the time trend, ${\beta}_{1},{\beta}_{2},{\beta}_{3},{\beta}_{4}$ represents vectors of coefficients to be estimated, and $\text{ε}$ is an error term, which is assumed to be normally distributed.

Secondly, through the regression result, I will test the overall hypothesis by using F-test and test the individual hypothesis by using T-test. The adjusted R-squared value will also show how well the data fit the regression model (the goodness of fit). 

### Assumption Testing
To validate our multiple regression model, below are the list of assumptions we need to check:

- #### Linearity
Definition: There should be a linear relationship between the independent variables and the dependent variable.
Result: 

```python
# Get the list of columns
x_columns = select_col_cleaned.columns[2:6]

# Create subplots for each pair of x and y
fig, axes = plt.subplots(nrows=1, ncols=len(x_columns), figsize=(20, 4), sharey=True)

# Loop through each pair of x and y
for i, col in enumerate(x_columns):
    x = select_col_cleaned[col]
    y = select_col_cleaned["collect"]
    
    # Calculate the correlation coefficient
    correlation_coefficient = x.corr(y)
    
    # Plot scatter plots
    axes[i].scatter(x, y)
    axes[i].set_title(f'{col} (Correlation: {correlation_coefficient:.2f})')

    axes[i].set_xlabel(col)
    if i == 0:
        axes[i].set_ylabel('NPI collection rate (%)')
    
    # Plot the diagonal line (perfect correlation) based on correlation direction
    min_val_x, max_val_x = min(x), max(x)
    min_val_y, max_val_y = min(y), max(y)
    if correlation_coefficient > 0:
        axes[i].plot([min_val_x, max_val_x], [min_val_y, max_val_y], linestyle='--', color='red', label='Positive Correlation')
    else:
        axes[i].plot([min_val_x, max_val_x], [max_val_y, min_val_y], linestyle='--', color='red', label='Negative Correlation')
    
    axes[i].legend()

plt.show()
```

- ####  Normality of the Error Terms
The error terms (residuals) are normally distributed, and thus satisfy the normality assumption.

```python
# Use statsmodels to estimate the residuals 

fig = plt.figure(constrained_layout=True)
gs = GridSpec(1, 2, figure=fig)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

df_lm = ols("collect ~ cpi+ecogrow+gdp+unemploy+month2+lagcpi", data=df).fit()
residuals = df['collect'] - df_lm.predict()
sm.qqplot(residuals, line='45', fit=True, ax=ax1) 
ax1.set_title('QQ Plot')

ax2.hist(residuals)
ax2.set_title('Histogram of the Residuals')
ax2.set(xlabel = 'Residuals')
ax2.set(ylabel = 'Frequency')
plt.show()
```

### 3.2 Bayesian estimation model

I sought to use one economic component to forecast future collection rates after testing correlations between economic factors and the Taiwan National Pension Insurance collection rate. I did not apply the Bayesian logic to CPI and GDP per capita because of the low t-value for the preceding regression. Due to its statistically significant effect on the collection rate, I opted to select just one independent variable "economic growth" from the model.

Testing whether we can predict monthly collection rate of National Pension Insurance using  𝑡−1  economic growth rate. 
Let our likelihood be described by:

$$y_{t+1} = \theta_y + \beta_e e_t + \sigma \varepsilon_{t+1}$$

with $\varepsilon_{t+1} \sim N(0, 1)$

We now specify priors over the mean of the collection rate $\theta_y$, economic growth rate $\beta_e$, and standard deviation $\sigma$:

- $\theta_y \sim N(56.26, 7.37)$
- $\beta_e \sim N(-7.8219, 26)$
- $\sigma \sim U(0.001, 10.0)$

```python
# Build a pymc3 model 
m = pm.Model()

# The model is used within a context manager (m)
# Insert us into a place where m is what everything will be operating on so everything can reference m 
# and that's where these variables and everything will get stored
with m:
    # Observed data
    # Specify our 'collect' variable as 'collect' and 'ecogrow' variable as eco_grow
    # The string in the first argument is going to internally identify each of these variables
    collect = pm.Data("collect", df["collect"])
    eco_growth = pm.Data("ecogrow", df["ecogrow"])

    # Priors
    # Assume each takes the normal distribution or uniform distribution with different mean and standard deviation
    theta_y = pm.Normal("theta_y", 56.26, 7.37)
    beta_e = pm.Normal("beta_e", -7.8219, 26)
    sigma = pm.Uniform("sigma", 0.001, 10.0)

    # Likelihood
    obs_ys = pm.Normal("obs_ys", theta_y + beta_e*eco_growth, sigma, observed=collect)
```

```python
# Draw 2000 samples and do manto carlo simulation with 2500 tuning period from our posterior and store them in a variable called trace
with m:
    trace = pm.sample(2000, tune=2500)
```

```python
# Mean of the posterior is right in the 60-61 range
fig, ax = plt.subplots(2, 2, figsize=(14, 8))

ax[0, 0].hist(trace["theta_y"], bins=50)
ax[0, 0].set_title("theta_y")
ax[0, 1].hist(trace["beta_e"], bins=50)
ax[0, 1].set_title("beta_e")
ax[1, 0].hist(trace["sigma"], bins=50)
ax[1, 0].set_title("sigma")
```

```python
# Mean of the posterior is 60.21
# If ecogrow = 0, mean of collection rate will be 60.21
np.mean(trace["theta_y"])
```

```python
# The standard deviation of the posterior is 0.93, which seems to be a normal distribution
np.std(trace["theta_y"]) 
```

```python
# 95 percent equal tailed credible set will go from 58.4% to 62.09%
# There are 95% possibility that the mean collection rate is between 58.4% to 62.09%
np.quantile(trace["theta_y"], [0.025, 0.975])  
```

```python
np.corrcoef(trace["theta_y"], trace["beta_e"])[0, 1]
```

```python
plt.scatter(trace["theta_y"], trace["beta_e"], alpha=0.4) 
# The economic growth rate is negatively correlated to the collection rate.
```

### Estimation results
After testing several models, I discovered that incorporating economic growth produced a more useful outcome. The above figure depicts the mean of the collection rate as being centered around 60%. Utilizing quantile to this distribution, I received [58.4001, 62.0924], meaning 95% of collection rate will be in this set under current economic growth rate. 

Finally, we can also conclude that even a better macroeconomic environment will still result in a decreasing collection rate of National Pension Insurance. 

* against our expectation that as the economic growth rate increases, the monthly collection rate will go down

## <span style="color:blue">Section 04. Conclusion</span> 
Holding all other factors constant, the findings point to a negative link between economic factors and the collection rate that is statistically significant at the 5% level. The reason why they are negatively correlated may be because one of the biggest properties lies in the pension policy: Paying premium is not mandatory since this pension is aimed to design for those who are unemployed or economically disadvantaged. Hence, this will provide citizens opportunities to choose their investment products other than national pension insurance, in order to ensure they can enjoy their elderly life at the basic economic level.

With better economic development, citizens do not have to worry they may lose their job in the future, and thus do not have the demand to insure the national pension insurance. Also, there are lots of alternatives to pensions in the financial market. While the national pension is only at the bottom of Taiwan’s social security system, and it can only provide people with the most basic living security, the incentives the national pension has are limited. That is, citizens with spare income may choose to buy stocks or bonds instead of national pension insurance.

The study is limited by the absence of regional series data on the collection rate of the National Pension Insurance, which might be Taiwan's panel data. We could also examine the subject by taking into account additional intersectional elements, such as gender, age, education, and the government's budget for promoting National Pension Insurance, in addition to the macroeconomic ones.
